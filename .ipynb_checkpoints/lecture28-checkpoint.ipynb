```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import sparse
from scipy.sparse import linalg as spla
import spectral
```

## Eigenvalue Problems

```python
N = 256
x_basis = spectral.Chebyshev(N)
x = x_basis.grid()
domain = spectral.Domain([x_basis])
```

```python
C = x_basis.convert_TU()
D = x_basis.derivative_TU()
Z = sparse.csr_matrix((N, N))
```

```python
M = sparse.csr_matrix((2*N+2, 2*N+2))
M[N:2*N, :N] = C

L = sparse.bmat([[D, -C],
                 [Z,  D]])

i = np.arange(N)
BC_rows = np.zeros((2, 2*N))
BC_rows[0, :N] = (-1)**i
BC_rows[1, :N] = (+1)**i

cols = np.zeros((2*N, 2))
cols[  N-1, 0] = 1
cols[2*N-1, 1] = 1

corner = np.zeros((2, 2))

L = sparse.bmat([[L, cols],
                 [BC_rows, corner]])
```

Let's do a dense solve to get all the eigenvalues of this problem.

```python
# :(
L = L.A
M = M.A
```

```python
values, vectors = scipy.linalg.eig(-L, M)

finite = np.isfinite(values)
values = values[finite]
vectors = vectors[:, finite]

order = np.argsort(values.real)
values = values[order]
vectors = vectors[:, order]

om = np.sqrt(values)
```

```python
i = np.arange(len(om))+1
plt.figure()
plt.scatter(i, om)
plt.scatter(i, np.pi/2*i, marker='x')
```

```python
error = np.abs(om - np.pi/2*i)
plt.figure()
plt.plot(error)
plt.yscale('log')
```

Rule of thumb:
 - about half eigenvalues will be accurate (well-resolved)
 - other half will be unaccurate

```python
def wave_evp(N):
    x_basis = spectral.Chebyshev(N)
    x = x_basis.grid()
    domain = spectral.Domain([x_basis])
    
    C = x_basis.convert_TU()
    D = x_basis.derivative_TU()
    Z = sparse.csr_matrix((N, N))
    
    M = sparse.csr_matrix((2*N+2, 2*N+2))
    M[N:2*N, :N] = C

    L = sparse.bmat([[D, -C],
                     [Z,  D]])

    i = np.arange(N)
    BC_rows = np.zeros((2, 2*N))
    BC_rows[0, :N] = (-1)**i
    BC_rows[1, :N] = (+1)**i

    cols = np.zeros((2*N, 2))
    cols[  N-1, 0] = 1
    cols[2*N-1, 1] = 1

    corner = np.zeros((2, 2))

    L = sparse.bmat([[L, cols],
                     [BC_rows, corner]])
    
    # :(
    L = L.A
    M = M.A
    
    values, vectors = scipy.linalg.eig(-L, M)

    finite = np.isfinite(values)
    values = values[finite]
    vectors = vectors[:, finite]

    order = np.argsort(values.real)
    values = values[order]
    vectors = vectors[:, order]

    om = np.sqrt(values)
    
    return om
```

```python
om_256 = wave_evp(256)
om_512 = wave_evp(512)
```

## Sparse Eigenvalue Solves

```python
N = 4096
x_basis = spectral.Chebyshev(N)
x = x_basis.grid()
domain = spectral.Domain([x_basis])

C = x_basis.convert_TU()
D = x_basis.derivative_TU()
Z = sparse.csr_matrix((N, N))

M = sparse.csr_matrix((2*N+2, 2*N+2))
M[N:2*N, :N] = C

L = sparse.bmat([[D, -C],
                 [Z,  D]])

i = np.arange(N)
BC_rows = np.zeros((2, 2*N))
BC_rows[0, :N] = (-1)**i
BC_rows[1, :N] = (+1)**i

cols = np.zeros((2*N, 2))
cols[  N-1, 0] = 1
cols[2*N-1, 1] = 1

corner = np.zeros((2, 2))

L = sparse.bmat([[L, cols],
                 [BC_rows, corner]])
L = -L
```

We need to make a linear operator that calculates $(L-\sigma^2M)^{-1}.M$.

```python
sigma = (3*np.pi/2) + 0.01

L_si = L - sigma**2 * M
solver = spla.splu(L_si)
def matvec(X):
    return solver.solve(M @ X)
```

```python
LO = spla.LinearOperator(dtype=L.dtype, shape=L.shape, matvec=matvec)
```

```python
LO
```

```python
evals, evecs = spla.eigs(LO, k=1, which='LM')
```

```python
evals
```

```python
np.sqrt(sigma**2 + 1/evals) - 3*np.pi/2
```
